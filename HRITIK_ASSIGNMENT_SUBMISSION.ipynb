```python
#IMPORTING IMPORTANT LIBRARIES FOR MODEL TRAINING AND TEST
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
import xgboost 
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
#LR,DTC,RFC,XGB
```

```python
#LABEL ENCODING
label_encoders = {}
df_columns = df.columns  

for i in df_columns:
    label_encoders[i] = LabelEncoder()
    df[i] = label_encoders[i].fit_transform(df[i])
```

```python
#STORING INDEPENDENT DATASET COLUMNS
X=df.drop(['status'],axis=1)
```

```python
#PREVIEWING TARGET COLUMN COUNTS
df.groupby(['status']).count()
```

```python
#STORING DEPENDENT DATASET COLUMN
y=df['status']
```

```python
#INDEPENDENT DATASET COLUMNS
X
```

```python
#DEPENDENT DATASET COLUMNS
y
```

```python
#SPLITING DATASET INTO TRAIN AND TEST SPLIT
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
#USING SMOTE LIBRARY FOR BALANCING DATASET BETWEEN MINORITY AND MAJORITY INPUTS
counter=Counter(y_train)
print('Before sampling',counter)
smt=SMOTE()
X_train_sm,y_train_sm= smt.fit_resample(X_train,y_train)
counter=Counter(y_train_sm)
print('After sampling',counter)
```

```python
#DIFFERENT CLASSIFICATION MACHINE LEARNING ALGORITHMS 
LR = LogisticRegression()
DTC = DecisionTreeClassifier()
RFC = RandomForestClassifier()
xgb_cl=xgboost.XGBClassifier()
```

```python
#CHECKING THE BEST ALGORITHM BY CHECKING THE BEST ACCURACY SCORE, PRECISION SCORE, RECALL SCORE AND F1 SCORE
li = [LR,DTC,RFC,xgb_cl]
q = {}
for i in li:
    i.fit(X_train,y_train)
    ypred = i.predict(X_test)
    print("Accuracy score for",i,":",accuracy_score(y_test,ypred)*100)
    print("Precision score for",i,":",precision_score(y_test,ypred)*100)
    print("Recall score for",i,":",recall_score(y_test,ypred)*100)
    print("F1 score for",i,":",f1_score(y_test,ypred)*100)
    print("*****************************************************************")
```

### AFTER GOING THROUGH THE ABOVE OBSERVATIONS, THE BEST MACHINE LEARNING ALGORITHM FOR THE GIVEN DATASET WOULD BE XGBOOSTCLASSIFIER WITH 98.31% OF ACCURACY SCORE, 87.67% OF PRECISION SCORE, 89.05% OF RECALL SCORE AND 88.46% OF F1 SCORE.
